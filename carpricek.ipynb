```python
car=car[car['year_of_manufacture'].apply(lambda x: pd.to_numeric(x, errors='coerce')).notnull()]
```

```python
car['car_price_in_rupees'] = car['car_price_in_rupees'].astype(str)
car['car_price_in_rupees'] = car['car_price_in_rupees'].str.replace('₹', '')
car['car_price_in_rupees'] = car['car_price_in_rupees'].str.replace('Lakh', '')
car['car_price_in_rupees'] = pd.to_numeric(car['car_price_in_rupees'], errors='coerce') * 100000
car['car_price_in_rupees']
```

```python
# Replace NaN with 0 and convert to integers
car['car_price_in_rupees'] = car['car_price_in_rupees'].fillna(0).astype(int)
```

```python
car.info()
```

```python
car['kms_driven'] = car['kms_driven'].astype(str)
car['kms_driven'] = car['kms_driven'].str.split(' ').str.get(0).str.replace(',', '')
```

```python
car['kms_driven']=car['kms_driven'].astype(int)
```

```python
X=car.drop(columns='car_price_in_rupees')
y=car['car_price_in_rupees']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe=OneHotEncoder()
ohe.fit(X[['car_name','city','fuel_type']])
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['car_name','city','fuel_type']),remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
# Assuming 'pipe' is your pipeline, and X_train, y_train are your training data
pipe.fit(X_train, y_train)

# Now you can use the predict method
y_pred = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
scores=[]
for i in range(1000):
  X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=i)
  lr=LinearRegression()
  pipe=make_pipeline(column_trans,lr)
  pipe.fit(X_train,y_train)
  y_pred=pipe.predict(X_test)

  scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
```
